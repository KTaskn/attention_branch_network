```python
import torchvision
dataset = iter(torchvision.datasets.CIFAR100("/tmp", download=True))
```

```python
image, _ = next(dataset)
display(image)

image, _ = next(dataset)
display(image)
```

```python
import torch
from models.cifar import resnet

PRETRAINED = "./pretrained/model_best.pth.tar"

checkpoint = torch.load(PRETRAINED)

model = resnet(
                    num_classes=100,
                    depth=110
                )
model = torch.nn.DataParallel(model).cuda()
model.load_state_dict(checkpoint['state_dict'])
```

```python
import torchvision.transforms as transforms
transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
    ])

resize = transforms.Resize((64, 64))
```

```python
for idx, (image, _) in enumerate(dataset):
    display(resize(image))
    image = torch.stack([
        transform(image)
    ]).cuda()

    _, outputs, attention = model(image)

    display(resize(transforms.functional.to_pil_image(attention[0])))
    
    if idx > 10:
        break
```
